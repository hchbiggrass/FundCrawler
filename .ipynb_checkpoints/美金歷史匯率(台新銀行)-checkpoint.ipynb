```python
# 美金歷史匯率(台新銀行)
'''資料來源為台新銀行，從2019年11月18日到2021/04/07
'''

from selenium import webdriver
import pandas as pd
import time
import gc

# 讀取最新資料日期
df = pd.read_csv('美金歷史匯率(台新銀行).csv')
Latest = df.iloc[-1,0]
del df
gc.collect()

data = {}

option = webdriver.ChromeOptions()
option.add_experimental_option('excludeSwitches', ['enable-automation'])
driver = webdriver.Chrome(options=option)
driver.maximize_window()

url = 'https://www.taishinbank.com.tw/TSB/personal/deposit/lookup/history/'
driver.get(url)

driver.find_element_by_css_selector('#ts-transac-comp-1-tab6 > div > div > div > div > form > div.ts-transac-comp-29 > ul > li:nth-child(3) > div > ul > li > div > div.select-block.select-taishin > div > button').click()
time.sleep(0.2)
driver.find_element_by_css_selector('#ts-transac-comp-1-tab6 > div > div > div > div > form > div.ts-transac-comp-29 > ul > li:nth-child(3) > div > ul > li > div > div.select-block.select-taishin > div > div > ul > li:nth-child(8) > a').click()
time.sleep(0.2)
RawDataString = driver.find_element_by_css_selector('#ts-transac-comp-1-tab6 > div > div > div > div > form > div.ts-comp-30 > div > table > tbody').text
RawData = RawDataString.split('\n')


for i in range(len(RawData)):
    data[i] = RawData[i].split(' ')

driver.close()
```

```python
# 將資料寫入csv
df = pd.DataFrame(data).transpose()
df.rename(columns={0:'日期' , 1:'即期買入' , 2:'即期賣出'}, inplace=True)
df = df.sort_values(by=['日期'], ascending=True)

# 篩選資料
df = df[df.日期 > Latest]

# 寫入csv
df.to_csv('美金歷史匯率(台新銀行).csv', header=0, index=0, mode='a')
```
