```python
#美金歷史匯率(比率網)
'''資料來源為比率網，從2006年1月2日到2021/02/02
'''
#資料為研究用，投資用要參考台新銀行
#原因是數字不知道是從哪裡來的，且跟台灣銀行和台新銀行都不同，難以直接比較

from selenium import webdriver
import pandas as pd
from tqdm import tqdm
import gc

#讀取最新資料日期
df = pd.read_csv('美金歷史匯率(比率網).csv')
Latest = df.iloc[-1,0]
del df
gc.collect()

option = webdriver.ChromeOptions()
option.add_argument('--headless')
option.add_experimental_option('excludeSwitches', ['enable-automation'])
driver = webdriver.Chrome(options=option)
driver.maximize_window()

data={}
count=0

#抓資料
for page in tqdm(range(1,12)): #先觀察自己需要幾頁，或是寫排程每天跑
    url = 'https://historical.findrate.tw/his.php?c=USD&page=' + str(page)
    driver.get(url)
    for i in range(2,32):
        try:
            count += 1
            #日期
            date = driver.find_element_by_css_selector('#right > div.newslist > table > tbody > tr:nth-child(' + str(i) + ') > td:nth-child(1)').text

            #即期買入
            buy = driver.find_element_by_css_selector('#right > div.newslist > table > tbody > tr:nth-child(' + str(i) + ') > td:nth-child(4)').text

            #即期賣出
            sale = driver.find_element_by_css_selector('#right > div.newslist > table > tbody > tr:nth-child(' + str(i) + ') > td:nth-child(5)').text
            
            #如果網頁上的資料比最新資料還要新就抓
            if date > Latest:
                data[count] = [date, buy, sale]
            else:
                break

        
        except:
            continue

driver.close()
```

```python
#將資料寫入csv
df = pd.DataFrame(data).transpose()
df.rename(columns={0:'日期' , 1:'即期買入' , 2:'即期賣出'}, inplace=True)
df = df.sort_values(by=['日期'], ascending=True)

#寫入csv
df.to_csv('美金歷史匯率(比率網).csv', header=0, index=0, mode='a')
```
